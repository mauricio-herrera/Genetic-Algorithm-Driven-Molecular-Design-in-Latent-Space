### Sanitizando el listado de molÃ©culas para entrenar el VAE

```python
Chem.MolFromSmiles('COc1ccc(-c2[nH]cc(CO)[nH+]2)cc1',sanitize=True)
```

```python
from rdkit import Chem

def sanitize_smiles(smiles_list):
    valid_smiles = []
    for smi in smiles_list:
        mol = Chem.MolFromSmiles(smi)
        if mol is None:
            print(f"Invalid SMILES string: {smi}")
            continue
        
        try:
            Chem.Kekulize(mol, clearAromaticFlags=True)
            valid_smiles.append(Chem.MolToSmiles(mol, kekuleSmiles=True))
        except Chem.KekulizeException:
            print(f"Can't kekulize mol: {smi}")
        except Exception as e:
            print(f"Error processing SMILES {smi}: {e}")
    
    return valid_smiles

def read_smiles_from_file(input_file):
    with open(input_file, 'r') as f:
        smiles_list = [line.strip() for line in f.readlines()]
    return smiles_list

def write_smiles_to_file(smiles_list, output_file):
    with open(output_file, 'w') as f:
        for smi in smiles_list:
            f.write(smi + '\n')

if __name__ == "__main__":
    input_file = "train.txt"  # Cambia este nombre por el de tu archivo de entrada
    output_file = "sanitized_smiles.txt"  # Este serÃ¡ el archivo de salida con las SMILES vÃ¡lidas

    smiles_list = read_smiles_from_file(input_file)
    valid_smiles = sanitize_smiles(smiles_list)
    write_smiles_to_file(valid_smiles, output_file)

    print(f"Sanitized SMILES saved to {output_file}")
```

## Experiments with genetic algorithms. Fitness includes Tanimoto and Lipinski's Rule of Five properties 

The initial molecule is random but can be replaced by a particular one to generate the initial population. Some parameters must be adjusted: tanimoto_weight, ngen, cxpb, mutpb, toolbox.population(n=50), among others.

Usamos un modelo de VAE pre-entrenado "model.epoch-19" 

```python
import torch
import numpy as np
import pandas as pd
from deap import base, creator, tools, algorithms
from fast_bo.gen_latent import scorer
from rdkit import Chem
from rdkit.Chem import Draw, AllChem, rdMolDescriptors
import matplotlib.pyplot as plt
from fast_molvae.sample import load_model

# Asegurar reproducibilidad
torch.manual_seed(42)
np.random.seed(42)

# Cargar el modelo entrenado
model = load_model('./data/vocab.txt', './fast_molvae/vae_model/model.epoch-19')

# Generar un vector latente inicial en CPU y asegurar una molÃ©cula vÃ¡lida
def get_valid_initial_vector(model):
    while True:
        z = torch.randn(1, 56 // 2)
        smile = model.decode(z, z, False)
        if Chem.MolFromSmiles(smile):
            return z.squeeze(0).numpy(), smile

initial_vector, initial_smile = get_valid_initial_vector(model)
initial_mol = Chem.MolFromSmiles(initial_smile)
initial_fp = AllChem.GetMorganFingerprintAsBitVect(initial_mol, radius=2, nBits=1024)
initial_properties = scorer([initial_smile])
initial_fitness = sum(initial_properties[0] + initial_properties[1] + initial_properties[2])

# PonderaciÃ³n del Tanimoto en el fitness
tanimoto_weight = 0.01

# FunciÃ³n de evaluaciÃ³n (fitness)
def evaluate(individual, model):
    z = torch.tensor(individual).float().unsqueeze(0)
    smile = model.decode(z, z, False)
    mol = Chem.MolFromSmiles(smile)
    if mol:
        # Calculate Lipinski's Rule of Five properties
        num_h_donors = rdMolDescriptors.CalcNumHBD(mol)
        num_h_acceptors = rdMolDescriptors.CalcNumHBA(mol)
        mol_weight = rdMolDescriptors.CalcExactMolWt(mol)
        log_p = rdMolDescriptors.CalcCrippenDescriptors(mol)[0]

        # Rule of Five compliance score
        rule_of_five_compliance = sum([
            num_h_donors <= 5,
            num_h_acceptors <= 10,
            mol_weight <= 500,
            log_p <= 5
        ])
        
        properties = scorer([smile])
        property_score = sum(properties[0] + properties[1] + properties[2])
        mol_fp = AllChem.GetMorganFingerprintAsBitVect(mol, radius=2, nBits=1024)
        tanimoto_score = AllChem.DataStructs.FingerprintSimilarity(initial_fp, mol_fp)
        
        # Adjusted fitness combining property score and Tanimoto similarity
        adjusted_fitness = property_score + (tanimoto_weight * tanimoto_score * 100) + rule_of_five_compliance
        return (adjusted_fitness,)
    return (-np.inf,)

# ConfiguraciÃ³n y ejecuciÃ³n del algoritmo genÃ©tico
creator.create("FitnessMax", base.Fitness, weights=(1.0,))
creator.create("Individual", list, fitness=creator.FitnessMax)
toolbox = base.Toolbox()
toolbox.register("individual", tools.initIterate, creator.Individual,
                 lambda: initial_vector + np.random.normal(0, 0.1, size=len(initial_vector)))
toolbox.register("population", tools.initRepeat, list, toolbox.individual)
toolbox.register("evaluate", evaluate, model=model)
toolbox.register("mate", tools.cxTwoPoint)
toolbox.register("mutate", tools.mutGaussian, mu=0, sigma=0.1, indpb=0.2)
toolbox.register("select", tools.selTournament, tournsize=3)

population = toolbox.population(n=50)
halloffame = tools.HallOfFame(1)
stats = tools.Statistics(lambda ind: ind.fitness.values)
stats.register("avg", np.mean)
stats.register("std", np.std)
stats.register("min", np.min)
stats.register("max", np.max)

algorithms.eaSimple(population, toolbox, cxpb=0.5, mutpb=0.2, ngen=10, stats=stats, halloffame=halloffame, verbose=True)

# Guardar las mejores molÃ©culas y sus propiedades
best_individual = halloffame[0]
best_smile = model.decode(torch.tensor(best_individual).float().unsqueeze(0), torch.tensor(best_individual).float().unsqueeze(0), False)
best_properties = scorer([best_smile])
best_fitness = sum(best_properties[0] + best_properties[1] + best_properties[2])
best_mol = Chem.MolFromSmiles(best_smile)
best_fp = AllChem.GetMorganFingerprintAsBitVect(best_mol, radius=2, nBits=1024)
tanimoto_best = AllChem.DataStructs.FingerprintSimilarity(initial_fp, best_fp)

# Create a DataFrame to save SMILES, fitness, and Tanimoto coefficient
results_df = pd.DataFrame({
    "SMILES": [initial_smile, best_smile],
    "Fitness": [initial_fitness, best_fitness],
    "Tanimoto_with_initial": [1, tanimoto_best]  # 1 for initial with itself, comparison for best with initial
})
results_df.to_csv("molecules_with_properties.csv", index=False)
```

```python
# FunciÃ³n para graficar molÃ©culas
def plot_molecules(smiles_list, titles_list):
    mols = [Chem.MolFromSmiles(smile) for smile in smiles_list if smile]
    imgs = [Draw.MolToImage(mol) for mol in mols if mol]
    fig, axs = plt.subplots(1, len(imgs), figsize=(10 * len(imgs), 6))
    if len(imgs) == 1:
        axs = [axs]  # Ensure axs is iterable
    for ax, img, title in zip(axs, imgs, titles_list):
        ax.imshow(img)
        ax.set_title(title)
        ax.axis('off')
    plt.show()

plot_molecules([initial_smile, best_smile], [
    "Initial Molecule: Fitness {:.2f}".format(initial_fitness),
    "Best Molecule: Fitness {:.2f}, Tanimoto with Initial: {:.2f}".format(best_fitness, tanimoto_best)
])
```

```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit import DataStructs

# Leer el archivo con las molÃ©culas
df = pd.read_csv("molecules_with_properties.csv")

# Obtener las SMILES de la primera y la Ãºltima molÃ©cula
first_smile = df.iloc[0]['SMILES']
best_smile = df.iloc[-1]['SMILES']

# Crear objetos molÃ©cula de RDKit a partir de las SMILES
first_mol = Chem.MolFromSmiles(first_smile)
best_mol = Chem.MolFromSmiles(best_smile)

# Calcular las huellas digitales de Morgan (Fingerprints)
fp1 = AllChem.GetMorganFingerprintAsBitVect(first_mol, radius=2, nBits=1024)
fp2 = AllChem.GetMorganFingerprintAsBitVect(best_mol, radius=2, nBits=1024)

# Calcular el coeficiente de Tanimoto
tanimoto_coefficient = DataStructs.FingerprintSimilarity(fp1, fp2)

print(f"Coeficiente de Tanimoto entre la primera molÃ©cula y la mejor molÃ©cula: {tanimoto_coefficient:.3f}")
```
